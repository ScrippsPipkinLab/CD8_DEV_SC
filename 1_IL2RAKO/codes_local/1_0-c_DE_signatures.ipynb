```python
import pandas as pd
import csv
import os
from pathlib import Path
import numpy as np
import glob
```

```python
###----- old - new cluster names
order_file = '/media/pipkin/ROCKET-PRO/CD8_DEV_SC/1_IL2RAKO/1_Scanpy/0-all_Scanpy_out/0_sum/1_IL2RAKO_obs_louvainAvg.csv'
order_df = pd.read_csv(order_file)
louvain_new_ordered = order_df['louvain_new_label'].tolist()
louvain_old_ordered = [str(x) for x in order_df['louvain'].tolist()]
order_df
```

```python
de_dir = '/media/pipkin/ROCKET-PRO/CD8_DEV_SC/1_IL2RAKO/1_Scanpy/0-all_Scanpy_out/2_DE'

path_dict = {}
path_dict['general'] ='%s/KO_vs_WT/CD25KO/CD25KO_vs_CD25WT_differential.csv'%de_dir

percluster_de_dir = '%s/perCluster_KO_vs_WT'%de_dir
glob.glob("%s/*.csv"%percluster_de_dir)

for path in Path(percluster_de_dir).rglob('*CD25KO_vs_CD25WT_differential.csv'):
    path_str = str(path.absolute())
    if "P" not in str(path_str.split("/")[-3]): 
        path_dict[path_str.split("/")[-3]] = path_str
path_dict
```

```python
louvain_new_ordered, louvain_old_ordered
```

```python
t_o_padj_c = 0.1
t_o_log2fc_c = 1
t_o_padj_c_p1 = 0.01
t_o_log2fc_c_p1 = 2
t_o_padj_c_p2 = 0.1
t_o_log2fc_c_p2 = 0

gs_names = []
g_symbols = []
for key, value in path_dict.items():
    if key in louvain_old_ordered:
        label = louvain_new_ordered[louvain_old_ordered.index(key)]
    else:
        label = key
    print("-----")
    print(key,label)
    if label == 'P1':
        t_o_padj_c, t_o_log2fc_c = t_o_padj_c_p1, t_o_log2fc_c_p1
    if label == 'P2':
        t_o_padj_c, t_o_log2fc_c = t_o_padj_c_p2, t_o_log2fc_c_p2
    diff_df = pd.read_csv(value)
    diff_df_sig = diff_df[diff_df['t-test_overestim_var_padj'] <= t_o_padj_c]
    diff_df_sig = diff_df_sig[np.abs(diff_df_sig['t-test_overestim_var_logfc']) >= t_o_log2fc_c]
    up_list = diff_df_sig[diff_df_sig['t-test_overestim_var_logfc'] > 0]['gene_names'].tolist()
    dn_list = diff_df_sig[diff_df_sig['t-test_overestim_var_logfc'] < 0]['gene_names'].tolist()
    
    print(len(up_list), len(dn_list))
    if len(up_list) > 10:
        gs_names += ["%s_Il2raKO_up"%label for x in up_list]
        g_symbols += up_list
    if len(dn_list) > 10:
        gs_names += ["%s_Il2raKO_down"%label for x in dn_list]
        g_symbols += dn_list
    

gs_df = pd.DataFrame({"gs_name": gs_names, "gene_symbol":g_symbols})
gs_df.to_csv("%s/D6-5_IL2_dependent_mm_sigs.csv"%de_dir, index=False)
```

```python
gs_df_sum = gs_df.groupby('gs_name').count()
gs_df_sum.columns = ['gene_number']
gs_df_sum.to_csv("%s/D6-5_IL2_dependent_mm_sigs_sum.csv"%de_dir)
gs_df_sum
```
