## 1. GSEA analysis for each set of comparison

```r
wk.dir <- '/media/pipkin/Yolanda/Exp391_Acute-Chronic_SC/1_2_scVelo/2_GSEA'
setwd(wk.dir)

gs.all.file <- '/media/pipkin/Yolanda/Exp391_Acute-Chronic_SC/z_References/GSEA/Combined_GEO_GSEA_signatures.csv'

de.out.dir <- '/media/pipkin/Yolanda/Exp391_Acute-Chronic_SC/1_2_scVelo/1_diff_kenetics'
de.files <- list.files(de.out.dir, pattern="*diff_kenetics_genes.csv", full.names=TRUE,recursive=TRUE)
de.files <- de.files[!grepl("all", de.files)] # Remove comparisons for all groups bc not fitting for X v.s. Y situation
```

```r
# Pick up where it's left* <br>
completed.name <- paste(wk.dir, "/", "completed.csv", sep="")
completed.file <- read_csv(completed.name)
completed <- completed.file$completed
de.files <- de.files[!(de.files %in% completed)]
```

```r
completed <- c()
for (file.i in de.files){
    file.i.outpath <- gsub(de.out.dir, "", file.i)
    file.i.outpath <- gsub(".csv", "", file.i.outpath)
    file.i.outpath <- gsub("^/", "",file.i.outpath)

    file.i.simp.name <- tail(unlist(strsplit( file.i.outpath, "/")), 1)
    file.i.simp.name <- gsub("_diff_kenetics_genes", "", file.i.simp.name)

    dir.create(file.path(wk.dir, file.i.outpath), recursive=TRUE, showWarnings=FALSE)
    setwd(file.path(wk.dir, file.i.outpath))

    # Calculate log2 score
    df.i <- read_csv(file.i) %>% mutate(sqrtScore = score/abs(score)*sqrt(abs(score)))
    file.i.new <- paste(file.i.simp.name, "_diff_kenetics_genes_sqrt.csv", sep="")
    write_csv(df.i, file.i.new)

    # Run GSEA analysis
    useGroup <- "sqrtScore"
    outName <- file.i.simp.name
    GSEA_analysis(file.i.new, useGroup, outName, gs.all.file, 0)
    
    # Write completed files
    completed <- c(completed, file.i)
    completed.df <- data.frame("completed" = completed)
    write.csv(completed.df, completed.name, row.names=TRUE)
}
```
