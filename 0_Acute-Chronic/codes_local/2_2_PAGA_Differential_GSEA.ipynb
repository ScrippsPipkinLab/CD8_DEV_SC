## 1. GSEA analysis for each set of comparison

```r
wk.dir <- '/media/pipkin/Yolanda/Exp391_Acute-Chronic_SC/1_1_SCANPY_PAGA/all--numSlt-rmWTNAV/4_DE_GSEA'
setwd(wk.dir)

gs.all.file <- '/media/pipkin/Yolanda/Exp391_Acute-Chronic_SC/z_References/GSEA/Combined_GEO_GSEA_signatures.csv'

de.out.dir <- '/media/pipkin/Yolanda/Exp391_Acute-Chronic_SC/1_1_SCANPY_PAGA/all--numSlt-rmWTNAV/1_DifferentialAnalysis'
de.files <- list.files(de.out.dir, pattern="*differential.csv", full.names=TRUE,recursive=TRUE)

use.cols <- c("gene_names", "t-test_overestim_var_logfc", "t-test_overestim_var_padj")
```

```r
completed <- c()
for (file.i in de.files) {

    ###----- Create output names & output directory
    file.i.outpath <- gsub(de.out.dir, "", file.i)
    file.i.outpath <- gsub(".csv", "", file.i.outpath)
    file.i.outpath <- gsub("^/", "",file.i.outpath)

    file.i.simp.name <- tail(unlist(strsplit( file.i.outpath, "/")), 1)
    file.i.simp.name <- gsub("_differential", "", file.i.simp.name)
    file.i.simp.vec <- unlist(strsplit(file.i.simp.name, "_vs_"))
    file.j.simp.name <- paste(rev(file.i.simp.vec), collapse="_vs_")

    ###----- Find pairing comparison file
    file.j.full.path <- c()
    for (j in de.files){
        if (grepl(file.j.simp.name, j)){
            file.j.full.path <- c(file.j.full.path, j)
        }
    }

    ###----- If succesully paired, combine paired files & run GSEA analysis
    if (length(file.j.full.path) == 1){
        file.j <- file.j.full.path[1]
        # Create output directory
        dir.create(file.path(wk.dir, file.i.outpath), recursive=TRUE, showWarnings=FALSE)
        setwd(file.path(wk.dir, file.i.outpath))

        # Create df comtaining genes of both sides
        i.tb <- read_csv(file.i) %>% dplyr::select(one_of(use.cols))
        j.tb <- read_csv(file.j) %>% dplyr::select(one_of(use.cols)) %>% 
                                        mutate(`t-test_overestim_var_logfc` = - `t-test_overestim_var_logfc`)
        ij.tb <- i.tb %>% bind_rows(j.tb)
        ij.name <- paste(file.i.simp.name, "_bothSides.csv", sep="")
        write_csv(ij.tb, ij.name)

        # Run GSEA analysis
        useGroup <- "t-test_overestim_var_logfc"
        outName <- file.i.simp.name
        GSEA_analysis(ij.name, useGroup, outName, gs.all.file, 1)

    } else {
        error.name <- paste(wk.dir, "/", file.i.simp.name, "_file_pairing_error.csv", sep="")
        error.df <- data.frame("input1" = c(file.i), "input2" = c(paste(file.j.full.path, collapse="|")))
        write.csv(error.df, error.name, row.names=FALSE)

    }
    completed.name <- paste(wk.dir, "/", "completed.csv", sep="")
    completed <- c(completed, file.i)
    completed.df <- data.frame("completed" = completed)
    write.csv(completed.df, completed.name, row.names=TRUE)
}
```
