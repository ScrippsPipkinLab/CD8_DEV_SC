```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
out_dir = '/media/pipkin/ROCKET-PRO/CD8_DEV_SC/9_Figures/9_Meta/heatmap'
```

```python
use_genes = ['Zeb2','Med12l','Gzma','Klrg1','Bhlhe40','Runx3','Tbx21','Chd7','Nt5e','Il7r','Sell','Tcf7','Slamf6']

shChd7_file = '/media/pipkin/ROCKET-PRO/CD8_DEV_SC/5_Chd7_shRNA_RNA/1_DEseq2_merge_shChd7/Chd7_vs_Ctrl_gn.csv'
shChd7_df = pd.read_csv(shChd7_file, index_col=0)

shBrg1_file = '/media/pipkin/ROCKET-PRO/CD8_DEV_SC/4_Runx3_Brg1_Meta/1_Brg1_RNA/1_DE/shBrg1_vs_shCd19_gn.csv'
shBrg1_df = pd.read_csv(shBrg1_file, index_col=0)

shChd7_df_use = shChd7_df.loc[use_genes][['log2FoldChange']]
shBrg1_df_use = shBrg1_df.loc[use_genes][['log2FoldChange']]

shChd7_df_use.columns = ['shChd7_vs_Ctrl']
shBrg1_df_use.columns = ['shBrg1_vs_Ctrl']

plot_df = shChd7_df_use.join(shBrg1_df_use)
plt.figure(figsize=(2,8))
sns.heatmap(plot_df,vmin=-2, vmax=2, cmap='vlag')
plt.savefig(out_dir + '/shBrg1-shChd7_heatmap.png')
```

```python
chd7ko_file = '/media/pipkin/ROCKET-PRO/CD8_DEV_SC/5_Chd7_KO_RNA/0_salmon_DE_out/2_DEseq_compiled/DESeq2_log2FoldChange.csv'
chd7ko_df = pd.read_csv(chd7ko_file, index_col=0)[['KO_TE-WT_TE','KO_MP-WT_MP','KO_EE-WT_EE']].loc[use_genes]


plt.figure(figsize=(2,8))
sns.heatmap(chd7ko_df,vmin=-2, vmax=2, cmap='vlag')

plt.savefig(out_dir + '/Chd7KO_heatmap.png')
```
