```python
def adata_louvain_name_convert(inputAdata, orderDf, louvainNewOrdered, louvainOldOrdered):
    inputAdata.obs["louvain_celltype"] = ["%s_%s"%(x,y) for index, (x,y) in enumerate(zip(inputAdata.obs['louvain'], inputAdata.obs['cell_type']))]
    inputAdata.obs["celltype_general"] = [x[0] for x in inputAdata.obs['cell_type']]
    
    #--- Match old & new louvain levels
    inputAdata.obs['louvain_old'] = inputAdata.obs['louvain']
    inputAdata.uns['louvain_old_colors'] = inputAdata.uns['louvain_colors']
    inputAdata.obs['louvain'] = [ louvainNewOrdered[louvainOldOrdered.index(int(x))] for x in inputAdata.obs['louvain'] ]
    category_order = orderDf.sort_values('louvain')['louvain_new_label'].tolist()
    category_order = [x for x in category_order if x in np.unique(inputAdata.obs['louvain'])]
    inputAdata.obs['louvain'] = pd.Categorical(inputAdata.obs['louvain'].tolist(), categories=category_order) # The key is to set categories according to original order...

    inputAdata.obs['Infection type'] = [ 'Armstrong' if 'A' in x else 'Clone13' if 'C' in x else 'Naive' for x in inputAdata.obs['cell_type']]
    inputAdata.obs['cell_id'] = inputAdata.obs.index.tolist()
```

## Differential driver genes

```python
plot_genes_cl13 = ['Il2ra', 'Lag3', 'Tox', 'Id2']
plot_genes_arm = ['Sell', 'Zeb2', 'Ccr7', 'Ctla4']
plot_genes_both = ['Tcf7', 'E2f1']
```

```python
arm_velocity = pd.DataFrame(adata_arm.layers['velocity'])
arm_velocity.index = adata_arm.obs.index
arm_velocity.columns = adata_arm.var.index

cl13_velocity = pd.DataFrame(adata_cl13.layers['velocity'])
cl13_velocity.index = adata_cl13.obs.index
cl13_velocity.columns = adata_cl13.var.index

arm_expr = pd.DataFrame.sparse.from_spmatrix(adata_arm.X).sparse.to_dense()
arm_expr.index = adata_arm.obs.index
arm_expr.columns = adata_arm.var.index

cl13_expr = pd.DataFrame.sparse.from_spmatrix(adata_cl13.X).sparse.to_dense()
cl13_expr.index = adata_cl13.obs.index
cl13_expr.columns = adata_cl13.var.index
```

```python
v_avg_df = pd.DataFrame(index=louvain_new_ordered)
e_avg_df = pd.DataFrame(index=louvain_new_ordered)
for gene_i in plot_genes_cl13 + plot_genes_arm + plot_genes_both:
    gene_i_arm_v_df = pd.DataFrame(arm_velocity[gene_i])
    gene_i_arm_v_df['louvain'] = adata_arm.obs['louvain']
    gene_i_arm_v_avg = gene_i_arm_v_df.groupby('louvain').mean()
    gene_i_arm_v_avg.columns = ["Arm_%s"%gene_i]

    gene_i_cl13_v_df = pd.DataFrame(cl13_velocity[gene_i])
    gene_i_cl13_v_df['louvain'] = adata_cl13.obs['louvain']
    gene_i_cl13_v_avg = gene_i_cl13_v_df.groupby('louvain').mean()
    gene_i_cl13_v_avg.columns = ["Cl13_%s"%gene_i]

    gene_i_arm_e_df = pd.DataFrame(arm_expr[gene_i])
    gene_i_arm_e_df['louvain'] = adata_arm.obs['louvain']
    gene_i_arm_e_avg = gene_i_arm_e_df.groupby('louvain').mean()
    gene_i_arm_e_avg.columns = ["Arm_%s"%gene_i]

    gene_i_cl13_e_df = pd.DataFrame(cl13_expr[gene_i])
    gene_i_cl13_e_df['louvain'] = adata_cl13.obs['louvain']
    gene_i_cl13_e_avg = gene_i_cl13_e_df.groupby('louvain').mean()
    gene_i_cl13_e_avg.columns = ["Cl13_%s"%gene_i]

    v_avg_df = v_avg_df.join(gene_i_arm_v_avg).join(gene_i_cl13_v_avg)
    e_avg_df = e_avg_df.join(gene_i_arm_e_avg).join(gene_i_cl13_e_avg)
```

```python
rcParams['figure.figsize'] = 3,6
sns.heatmap(v_avg_df.T, cmap='vlag', center=0)
```

```python
rcParams['figure.figsize'] = 3,6
sns.heatmap(e_avg_df.T, cmap='viridis_r')
```

```python
driver_genes = ['Uhrf1', 'Mcm4', 'Tfdp1', 'Parp1', 'Il2ra', 'Snrpd1', 'Trip13', 'Hells', 'Wdhd1', 'Timeless', 'E2f8']

act_velocity = pd.DataFrame(adata_act.layers['velocity'])
act_velocity.index = adata_act.obs.index
act_velocity.columns = adata_act.var.index

act_expr = pd.DataFrame.sparse.from_spmatrix(adata_act.X).sparse.to_dense()
act_expr.index = adata_act.obs.index
act_expr.columns = adata_act.var.index
```

```python
v_avg_df = pd.DataFrame(index=louvain_new_ordered)
e_avg_df = pd.DataFrame(index=louvain_new_ordered)
for gene_i in driver_genes:
    gene_i_act_v_df = pd.DataFrame(act_velocity[gene_i])
    gene_i_act_v_df['louvain'] = adata_act.obs['louvain']
    gene_i_act_v_avg = gene_i_act_v_df.groupby('louvain').mean()
    gene_i_act_v_avg.columns = [gene_i]

    gene_i_act_e_df = pd.DataFrame(act_expr[gene_i])
    gene_i_act_e_df['louvain'] = adata_act.obs['louvain']
    gene_i_act_e_avg = gene_i_act_e_df.groupby('louvain').mean()
    gene_i_act_e_avg.columns = [gene_i]

    v_avg_df = v_avg_df.join(gene_i_act_v_avg)
    e_avg_df = e_avg_df.join(gene_i_act_e_avg)
```

```python
rcParams['figure.figsize'] = 3,3
sns.heatmap(v_avg_df.T, cmap='vlag', center=0)
```
