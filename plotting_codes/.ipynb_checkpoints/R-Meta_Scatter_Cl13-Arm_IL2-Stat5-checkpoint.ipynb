```r
library(tidyverse)
library(ggrepel)
if ( ! exists("code_dir")) {
    code_dir <- getwd()
    base_dir <- gsub("/plotting_codes", "", code_dir)
    meta.dir <- file.path(base_dir, '/9_Figures/9_Meta')
}
```

```r
avg.e.file <- file.path(base_dir, '/0_Acute-Chronic/1_Scanpy/0_Scanpy_out_resampled/1_avg_expr/louvain_celltype_general_mean_scaled_expr.csv')
avg.e.df <- read_csv(avg.e.file)

il2.stat5.csv.name <- file.path(meta.dir, "Il2-stat5_direct_promote.csv")
il2.stat5.df <- read_csv(il2.stat5.csv.name)

avg.e.df$il2_stat5 <- ifelse(avg.e.df$gene_name %in% il2.stat5.df$gene_name, TRUE, FALSE)

anno.genes <- c("Tcf7", "Chd7", "Aff3", "Batf3", "Runx3")
avg.e.anno.df <- avg.e.df %>% filter(gene_name %in% anno.genes)
```

```r
scatter.plot <- ggplot(avg.e.df, aes(x=`6_A`, y=`6_C`, color=il2_stat5, alpha=il2_stat5, size=il2_stat5)) +
    geom_point() + 
    scale_color_manual(values=c( "black", "red")) +
    scale_alpha_manual(values=c(0.1, 0.6)) +
    scale_size_manual(values=c(1.5,1.5)) +
    
    geom_vline(xintercept=0) +
    geom_hline(yintercept=0) +
    xlim(c(-1, 1)) + ylim(c(-1,1)) +
    #geom_smooth(method=lm, se=FALSE, fullrange=TRUE) +
    
    geom_label_repel(data=avg.e.anno.df, size=12, alpha=1, force=80, max.iter=100000,
                          aes(label=gene_name)) +
    geom_point(data=avg.e.anno.df, pch=1, size=3, alpha=1) +
    
    theme(axis.line = element_line(colour = "black"),
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
    panel.border = element_blank(),
    panel.background = element_blank(),
    axis.title.x = element_blank(),
    axis.title.y = element_blank(),
    legend.position = "none")
scatter.plot
```
